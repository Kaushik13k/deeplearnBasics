```python
import numpy as np
from numpy import asarray

data = [[3, 3, 2, 1, 0],
        [0, 0, 1, 3, 1],
        [3, 1, 2, 2, 3],
        [2, 0, 0, 2, 2],
        [2, 0, 0, 0, 1]]
data = np.asarray(data)
data = data.reshape(1, 5, 5, 1) # get 5 x 5 

kernel = [[[[0]], [[1]], [[2]]],
          [[[2]], [[2]],[[0]]],
          [[[0]], [[1]], [[2]]]]

weights = [asarray(kernel), asarray([0.0])]
```

```python
print(data)
```

```python
print(kernel)
```

```python
from keras.models import Sequential
from keras.layers import Conv2D

# Create model
## KERNEL
model = Sequential()
model.add(Conv2D(1, (3,3), input_shape=(5, 5, 1)))
print(model.summary())
print('-----------------------------------------------------------------------------------------------------------------------')
model.set_weights(weights)
yhat = model.predict(data)
print('the output is :')
print('---------------')
for r in range(yhat.shape[1]):
    print([yhat[0, r, c, 0] for c in range(yhat.shape[2])])
```

```python
## Stride
model = Sequential()
model.add(Conv2D(1, (3,3), strides=(2,2), input_shape=(5, 5, 1)))
print(model.summary())
print('-----------------------------------------------------------------------------------------------------------------------')
model.set_weights(weights)
yhat = model.predict(data)
print('the output is :')
print('---------------')
for r in range(yhat.shape[1]):
    print([yhat[0, r, c, 0] for c in range(yhat.shape[2])])
```

```python
## Padding
model = Sequential()
model.add(Conv2D(1, (3,3), padding='same', input_shape=(5, 5, 1)))
print(model.summary())
print('-----------------------------------------------------------------------------------------------------------------------')
model.set_weights(weights)
yhat = model.predict(data)
print('the output is :')
print('---------------')
for r in range(yhat.shape[1]):
    print([yhat[0, r, c, 0] for c in range(yhat.shape[2])])
```

```python
## Max Pooling
from keras.layers import MaxPooling2D
model = Sequential()
model.add(Conv2D(1, (3,3), padding='same', input_shape=(5, 5, 1)))
model.add(MaxPooling2D((2, 2)))
print(model.summary())
print('-----------------------------------------------------------------------------------------------------------------------')
model.set_weights(weights)
yhat = model.predict(data)
print('the output is :')
print('---------------')
for r in range(yhat.shape[1]):
    print([yhat[0, r, c, 0] for c in range(yhat.shape[2])])
```

```python
## Average Pooling
from keras.layers import AveragePooling2D
model = Sequential()
model.add(Conv2D(1, (3,3), padding='same', input_shape=(5, 5, 1)))
model.add(AveragePooling2D((2, 2)))
print(model.summary())
print('-----------------------------------------------------------------------------------------------------------------------')
model.set_weights(weights)
yhat = model.predict(data)
print('the output is :')
print('---------------')
for r in range(yhat.shape[1]):
    print([yhat[0, r, c, 0] for c in range(yhat.shape[2])])
```

```python
# Flatten
from keras.layers import Flatten

model = Sequential()
model.add(Conv2D(1, (3,3), padding='same', input_shape=(5, 5, 1)))
model.add(AveragePooling2D((2, 2)))
model.add(Flatten())
model.summary()
```
